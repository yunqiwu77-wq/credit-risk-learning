```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

loan_data = pd.DataFrame({
    "Client": ["Alice","Bob","Charlie","David","Emma","Frank","Grace","Helen"],
    "Age":[25,42,37,29,51,45,33,27],
    "Income":[35000,72000,58000,41000,90000,68000,55000,39000],
    "Loan":[12000,25000,18000,8000,40000,28000,15000,10000],
    "Default":[0,0,1,0,1,0,1,0]
})

loan_data
```

```python
loan_data.shape
```

```python
loan_data.columns
```

```python
loan_data.describe()
```

```python
loan_data["Income"].hist(
    bins=5
)

plt.title("Income Distribution")

plt.show()
```

```python
loan_data["Loan"].hist(
    bins=5
)

plt.title("Loan Distribution")

plt.show()
```

```python
loan_data["Default"].value_counts()
```

```python
loan_data["Default"].value_counts().plot(
    kind="bar"
)

plt.title("Default Distribution")

plt.show()
```

```python
loan_data.sort_values(
    by="Loan",
    ascending=False
)
```

```python
plt.scatter(
    loan_data["Income"],
    loan_data["Loan"]
)

plt.xlabel("Income")

plt.ylabel("Loan")

plt.title("Income vs Loan")

plt.show()
```

```python
loan_data["DTI"] = (
    loan_data["Loan"]
    /
    loan_data["Income"]
)
```

```python
loan_data.sort_values(
    by="DTI",
    ascending=False
)
```
