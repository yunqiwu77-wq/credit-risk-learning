```python
import pandas as pd
import numpy as np

loan_data = pd.DataFrame({
    "Client": ["Alice", "Bob", "Charlie", "David", "Emma", "Frank"],
    "Age": [25, 42, np.nan, 29, 51, 35],
    "Income": [35000, 72000, 58000, np.nan, 90000, 90000],
    "Loan": [12000, 25000, 18000, 8000, 40000, 40000],
    "Default": [0, 0, 1, 0, 1, 1]
})

loan_data
```

```python
loan_data.isnull()
```

```python
loan_data.isnull().sum()
```

```python
loan_data.dropna()
```

```python
loan_data["Age"]=loan_data["Age"].fillna(loan_data["Age"].mean())
```

```python
loan_data
```

```python
loan_data["Income"]=loan_data["Income"].fillna(loan_data["Income"].mean())
```

```python
loan_data
```

```python
loan_data = pd.concat(
    [loan_data, loan_data.iloc[[0]]],
    ignore_index=True
)
```

```python
loan_data.duplicated()
```

```python
loan_data
```

```python
loan_data=loan_data.drop_duplicates()
```

```python
loan_data
```

```python
loan_data.loc[2, "Income"] = 1000000
```

```python
loan_data.describe()
```

```python
loan_data[loan_data["Income"]>200000]
```

```python
loan_data.to_csv("clean_loan_data.csv",index=False)
```
